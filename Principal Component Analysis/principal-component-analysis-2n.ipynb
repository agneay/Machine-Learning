```python
# Create the loading matrix from the question
loadings = pd.DataFrame({
    "PC1": [0.190, 0.544, 0.782, 0.365, 0.585, 0.394, 0.985, 0.520, 0.142],
    "PC2": [0.017, 0.020, -0.605, 0.294, 0.085, -0.273, 0.126, 0.402, 0.150],
    "PC3": [0.207, 0.204, 0.144, 0.585, 0.234, 0.027, -0.111, 0.519, 0.239]
}, index=[
    "Climate",
    "Housing",
    "Health",
    "Crime",
    "Transportation",
    "Education",
    "Arts",
    "Recreation",
    "Economy"
])
```

```python
print("Loading Matrix:\n")
print(loadings)

print("\nDominant Variable in Each Principal Component:\n")
```

```python
for pc in loadings.columns:
    dominant_variable = loadings[pc].abs().idxmax()
    loading_value = loadings.loc[dominant_variable, pc]
    print(f"{pc}: {dominant_variable} (Loading = {loading_value})")
```
