## Importing Required Libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

## Importing (Reading) Datasets:

```python
data = pd.read_csv("/kaggle/input/datasets/abhishek14398/salary-dataset-simple-linear-regression/Salary_dataset.csv",index_col=0)
```

## Exploring Dataset

```python
data.head()
```

```python
data.shape
```

## Checking for any null values in dataset

```python
data.isnull().sum()
```

## Assigning Dependent and Independent variables

```python
x = data.iloc[:,:1].values
y = data.iloc[:,1:2].values
```

## Splitting the dataset into Training and testing dataset

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## Fitting the model (Linear Regression)

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
print(y_pred)
print(y_test)
```

## Plot for training Dataset

```python
%matplotlib inline
plt.scatter(x_train,y_train,color='blue')
plt.plot(x_train,model.predict(x_train),color='red')
plt.title('Salary VS Experience (training set)')
plt.xlabel('Experience in Years')
plt.ylabel('Salary in Rupees')
plt.show()
```

## Plot for validation dataset

```python
%matplotlib inline
plt.scatter(x_test,y_test,color="blue")
plt.plot(x_train,model.predict(x_train),color='red')
plt.title("Salary VS Experience (Validation dataset)")
plt.xlabel("Salary in Rupees")
plt.ylabel("Experience in Years")
plt.show()
```
