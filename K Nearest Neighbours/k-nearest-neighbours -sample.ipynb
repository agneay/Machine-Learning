```python
import pandas as pd
from sklearn.model_selection  import train_test_split 
from sklearn.neighbors  import KNeighborsClassifier 
from sklearn.metrics  import classification_report 
from sklearn import metrics
import seaborn as sn
```

```python
data = pd.read_csv(r"/kaggle/input/datasets/organizations/uciml/breast-cancer-wisconsin-data/data.csv")
print(data.shape)
data.head()
```

## Checking for any null values in the dataset

```python
data.isnull().sum()
```

```python
data = data.drop(columns=["Unnamed: 32"])
```

```python
data.head()
```

## Assigning Dependent and independant values

```python
# Remove the ID column
data = data.drop(columns=["id"])

# Features
x = data.drop(columns=["diagnosis"])

# Target
y = data["diagnosis"]
```

## Splitting the dataset

```python
x_train, x_test, y_train, y_test = train_test_split (x, y, test_size=0.3, random_state=0)
# display(x train.shape, y_train. shape, x test.shape, y_test. shape)
```

```python
data.columns
```

## Preprocessing data from standard scaler

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

## Training Model using KNN

```python
results=[]
for i in [1,2,3,4,5] :
    model = KNeighborsClassifier(n_neighbors=i,metric='minkowski', p=2)
    model.fit(x_train, y_train)
    y_pred=model.predict(x_test)
    Accuracy_score=metrics.accuracy_score(y_test, y_pred)
    results.append(Accuracy_score)

print('KNN [ minkowski ]')
print('for n_neighbor=5 : ')
conf_mat=metrics.confusion_matrix(y_test, y_pred)
print('n Confusion Matrix : ', conf_mat)
print ('Accuracy Score : ', Accuracy_score)
print('Accuracy in Percentage : ',
int(Accuracy_score*100), '%')
print('\n', classification_report(y_pred, y_test) )
print(results)
```

```python
conf_mat=pd.crosstab(y_test, y_pred, rownames=['Actual' ],
colnames=['Predicted' ] )
sn.heatmap(conf_mat, annot=True).set(title='KNN [minkowski, neighbor=5 ]')
```
