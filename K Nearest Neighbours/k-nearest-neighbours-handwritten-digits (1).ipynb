```python
import os
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Function to convert 32x32 text image into a 1x1024 vector
def img2vector(filename):
    returnVect = np.zeros((1, 1024))
    with open(filename) as fr:
        for i in range(32):
            lineStr = fr.readline().strip()
            for j in range(32):
                returnVect[0, 32 * i + j] = int(lineStr[j])
    return returnVect

# Dataset paths
train_path = "/kaggle/input/datasets/agneaybnair/digits/trainingDigits"
test_path = "/kaggle/input/datasets/agneaybnair/digits/testDigits"

trainingFileList = sorted(os.listdir(train_path))

trainingMat = np.zeros((len(trainingFileList), 1024))
hwLabels = []

for i, fileName in enumerate(trainingFileList):
    label = int(fileName.split("_")[0])
    hwLabels.append(label)
    trainingMat[i, :] = img2vector(os.path.join(train_path, fileName))[0]

print("Training samples:", len(hwLabels))


knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(trainingMat, hwLabels)

testFileList = sorted(os.listdir(test_path))

testMat = np.zeros((len(testFileList), 1024))
testLabels = []

for i, fileName in enumerate(testFileList):
    label = int(fileName.split("_")[0])
    testLabels.append(label)
    testMat[i, :] = img2vector(os.path.join(test_path, fileName))[0]

print("Test samples:", len(testLabels))


predictions = knn.predict(testMat)


print("\nAccuracy:", accuracy_score(testLabels, predictions))

print("\nConfusion Matrix:")
print(confusion_matrix(testLabels, predictions))

print("\nClassification Report:")
print(classification_report(testLabels, predictions))


print("\nSample Predictions:")
for i in range(10):
    print(f"Actual: {testLabels[i]}   Predicted: {predictions[i]}")
```
